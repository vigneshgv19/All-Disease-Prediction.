#### GRADIO INPUT

```python
#### CREATING GRADIO INPUT #####
Input = []
Column_name = []
for x1 in Train.columns:
    Column_name.append(x1)
    a = []
    for x2 in Train[x1].unique():
        a.append(x2)
    x1 = gd.inputs.Dropdown(a,label = x1) ### Gradio input for each feature variable
    Input.append(x1)
len(Input),len(Column_name)
```
